1. Considere el sistema de ecuaciones diferenciales dado por:

$dx(t) = -\alpha x(t)dt + \sigma dB_t$

$dy(t) = x(t)dt + t(-\alpha x(t) + \sigma dB_t)$

**Parámetros**

```python
X0 = 0
Y0 = 0
alpha = 5
sigma = 3
N = 250
dt = 1/250
t = np.arange(0,1,dt)
```

**Esquema de Euler**
$$x(t_i) = -\alpha x(t_{i-1}) \Delta{t_i} + \sigma \Delta B_{t_i}$$
$$y(t_i) = x(t_{i-1})\Delta{t_i} + t_{i-1}x(t_i)$$

```python
X = np.zeros(N)
Y = np.zeros(N)
X[0] = X0
Y[0] = Y0
for i in range(1,N):
    dbt = np.random.normal()
    X[i] = X[i-1] - alpha*X[i-1]*dt + sigma*np.sqrt(dt)*dbt
    Y[i] = X[i-1] + X[i-1]*dt + t[i-1]*X[i]
```

```python
plt.plot(t, X, 'red')
plt.plot(t, Y, 'green')
plt.title('Simulación con el esquema de Euler')
plt.show()
```

2. Considere la EDELH Bidimensional dada por:
$$dx(t) = -x(t)dt + Kx(t)dB_t$$
Con valor inicial $x(0) = (1,0)^T$
$$dx_1(t) = -x_1(t)dt -\sqrt{2}x_2(t)dB_t$$
$$dx_2(t) = -x_2(t)dt + \sqrt{2}x_2(t)dB_t$$

**Parámetros**

```python
X10 = 1
X20 = 0
N = 250
dt = 1/250
t = np.arange(0,1,dt)
```

**Simulación con esquema de Euler**

```python
X1 = np.zeros(N)
X2 = np.zeros(N)
X1[0] = X10
X2[0] = X20
for i in range(1,N):
    dbt = np.sqrt(dt)*np.random.normal()
    X1[i] = X1[i-1]-X1[i-1]*dt - np.sqrt(2)*X2[i-1]*dbt
    X2[i] = X2[i-1]-X2[i-1]*dt + np.sqrt(2)*X1[i-1]*dbt
```

```python
plt.plot(t,X1)
plt.plot(t,X2)
plt.title('Simulación con esquema de Euler')
plt.show()
```

4. _Ruido coloreado_: Considere la ED dada por:

$$dx(t) = a x(t)dt + by(t)dt$$

con valor inicial $x(0) = x_0$ donde $y(t)$ es el ruido coloreado:

$$dy(t) = -\alpha y(t)dt + \sigma dB_t$$

con valor inicial $y(0) = y_0$

**Parámetros**

```python
N = 250
dt = 1/N
alpha = 1
sigma = 0.1
a = 0.6
b = 0.5
t = np.arange(0,1,dt)
X0 = 2
Y0 = 2
```

**Simulación con esquema de Euler**

```python
X = np.zeros(N)
Y = np.zeros(N)
X[0] = X0
Y[0] = Y0
for i in range(1,N):
    dbt = np.sqrt(dt)*np.random.normal()
    X[i] = X[i-1] + a*X[i-1]*dt + b*Y[i-1]*dt
    Y[i] = Y[i-1] - alpha*Y[i-1]*dt + sigma*dbt
```

```python
plt.plot(t, X, 'orange')
plt.plot(t, Y, 'blue')
plt.title('Ruido coloreado')
plt.show()
```
